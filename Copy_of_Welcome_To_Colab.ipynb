```
import os

# class mapping (update as needed)
class_map = {
    "Paper": 0,
    "Cardboard": 1,
    "Plastic": 2,
    "Metal": 3,
    "Glass": 4,
    "Trash": 5
}

# mapping each dataset name to its assigned ID
dataset_class_id = {
    "Paper-3": class_map["Paper"],
    "Cardboard-3": class_map["Cardboard"],
    "Plastic-1": class_map["Plastic"],
    "Metal-1": class_map["Metal"],
    "Glasses-1": class_map["Glass"],   # assuming GreenLens is glass
    "Trash-1": class_map["Trash"]
}

merged_path = "/content/GreenLense_Merged/train/labels"

# Fix class IDs in merged label files
for dataset_name, new_id in dataset_class_id.items():
    folder = f"/content/{dataset_name}/train/labels"
    if not os.path.exists(folder):
        continue
    for file in os.listdir(folder):
        file_path = os.path.join(folder, file)
        if os.path.isfile(file_path):
            with open(file_path, "r") as f:
                lines = f.readlines()
            new_lines = []
            for line in lines:
                parts = line.strip().split()
                if len(parts) > 0:
                    parts[0] = str(new_id)  # replace class index
                    new_lines.append(" ".join(parts) + "\n")
            # overwrite label file in merged folder
            merged_label_path = os.path.join(merged_path, file)
            with open(merged_label_path, "w") as f:
                f.writelines(new_lines)

print("âœ… All label indices corrected according to your 6-class mapping!")
```

```
import pandas as pd

def calculate_carbon_footprint(gpu_type, training_time_hours, carbon_intensity_gco2_kwh=500):
    """
    Calculates the carbon footprint of a machine learning training run.

    Args:
        gpu_type (str): The model of the GPU (e.g., 'T4', 'V100', 'A100').
        training_time_hours (float): The total training time in hours.
        carbon_intensity_gco2_kwh (float): The carbon intensity of the electricity grid
                                           in grams of CO2 per kWh. Defaults to 500.
    """
    # Power consumption in Watts for common GPUs (from the original script)
    gpu_power_data = {
        'A100': 400,
        'V100': 250,
        'T4': 70,
        'P100': 250,
        'K80': 300,
        'RTX 3080': 320,
        'RTX 3090': 350
    }

    # --- 1. Get Power Consumption ---
    if gpu_type.upper() in gpu_power_data:
        power_consumption_watts = gpu_power_data[gpu_type.upper()]
        print(f"Using default power consumption for {gpu_type.upper()}: {power_consumption_watts}W")
    else:
        print(f"Error: Power consumption for GPU '{gpu_type}' is not in our database.")
        print("Please add it to the 'gpu_power_data' dictionary in the script.")
        return

    # --- 2. Calculate Energy Consumption ---
    # Energy (kWh) = Power (W) * Time (h) / 1000
    energy_kwh = (power_consumption_watts * training_time_hours) / 1000

    # --- 3. Calculate Carbon Emissions ---
    # Emissions (gCO2) = Energy (kWh) * Carbon Intensity (gCO2/kWh)
    carbon_emissions_g = energy_kwh * carbon_intensity_gco2_kwh

    # --- 4. Display Results ---
    print("\n---")
    print("Estimating Carbon Emissions for your ML Model")
    print(f"GPU Type: {gpu_type.upper()}")
    print(f"Training Time: {training_time_hours} hours")
    print(f"Power Consumption: {power_consumption_watts}W")
    print(f"Carbon Intensity: {carbon_intensity_gco2_kwh} gCO2/kWh")
    print("\n" + "="*40)
    print(f"Estimated Carbon Emissions: {carbon_emissions_g:.2f} gCO2")
    print("="*40)
    print("---\n")
```

```
import time
import subprocess
import re

def get_gpu_name():
    """Automatically detects the GPU name using nvidia-smi."""
    try:
        result = subprocess.run(['nvidia-smi', '--query-gpu=name', '--format=csv,noheader,nounits'],
                                capture_output=True, text=True, check=True)
        # Clean up the output to get the clean GPU name (e.g., "Tesla T4")
        gpu_name = result.stdout.strip().split(' ')[-1]
        return gpu_name
    except (subprocess.CalledProcessError, FileNotFoundError):
        return "CPU" # Fallback if nvidia-smi is not found or fails

def calculate_ml_carbon_footprint(start_time, end_time, gpu_name=None, carbon_intensity_gco2_kwh=400):
    """
    Calculates the carbon footprint of a completed ML training run.

    Args:
        start_time (float): The start time from time.time().
        end_time (float): The end time from time.time().
        gpu_name (str): The name of the GPU used. If None, it will be auto-detected.
        carbon_intensity_gco2_kwh (float): Grid carbon intensity. Default is 400 (US average).

    Returns:
        dict: A dictionary containing the detailed results.
    """
    # --- 1. Determine GPU and Power ---
    if gpu_name is None:
        gpu_name = get_gpu_name()

    # Power consumption in Watts for common GPUs
    gpu_power_data = {
        'A100': 400, 'V100': 250, 'T4': 70, 'P100': 250, 'K80': 300,
        'RTX3080': 320, 'RTX3090': 350, 'A10G': 150
    }

    # Handle CPU case
    if gpu_name == "CPU":
        print("Warning: Running on CPU. Power consumption is highly variable and not estimated here.")
        return {"error": "Cannot estimate CPU power consumption."}

    power_consumption_watts = gpu_power_data.get(gpu_name)
    if power_consumption_watts is None:
        print(f"Error: Power consumption for GPU '{gpu_name}' is not in our database.")
        return {"error": f"Unknown GPU: {gpu_name}"}

    # --- 2. Calculate Time and Energy ---
    training_time_seconds = end_time - start_time
    training_time_hours = training_time_seconds / 3600.0

    # Energy (kWh) = Power (W) * Time (h) / 1000
    energy_kwh = (power_consumption_watts * training_time_hours) / 1000

    # --- 3. Calculate Carbon Emissions ---
    # Emissions (gCO2) = Energy (kWh) * Carbon Intensity (gCO2/kWh)
    carbon_emissions_g = energy_kwh * carbon_intensity_gco2_kwh

    # --- 4. Return Results as a Dictionary ---
    results = {
        "gpu_name": gpu_name,
        "training_time_hours": round(training_time_hours, 4),
        "power_consumption_watts": power_consumption_watts,
        "energy_kwh": round(energy_kwh, 6),
        "carbon_intensity_gco2_kwh": carbon_intensity_gco2_kwh,
        "estimated_emissions_gco2": round(carbon_emissions_g, 2)
    }
    return results

def print_footprint_report(results):
    """Prints a nicely formatted report from the results dictionary."""
    if "error" in results:
        print(f"Calculation failed: {results['error']}")
        return

    print("\n" + "="*50)
    print("         CARBON FOOTPRINT REPORT")
    print("="*50)
    print(f"  GPU Used:                {results['gpu_name']}")
    print(f"  Training Duration:       {results['training_time_hours']} hours")
    print(f"  Avg. Power Consumption:  {results['power_consumption_watts']} Watts")
    print(f"  Total Energy Consumed:   {results['energy_kwh']} kWh")
    print(f"  Grid Carbon Intensity:   {results['carbon_intensity_gco2_kwh']} gCOâ‚‚/kWh")
    print("-"*50)
    print(f"  ğŸŒ ESTIMATED EMISSIONS:  {results['estimated_emissions_gco2']} gCOâ‚‚e")
    print("="*50 + "\n")
```

```
import time
import subprocess
import re
import json
import platform
import uuid
import psutil
import csv
from datetime import datetime

def get_gpu_name():
    """Automatically detects the GPU name using nvidia-smi."""
    try:
        result = subprocess.run(['nvidia-smi', '--query-gpu=name', '--format=csv,noheader,nounits'],
                                capture_output=True, text=True, check=True)
        gpu_name = result.stdout.strip().split(' ')[-1]
        return gpu_name
    except (subprocess.CalledProcessError, FileNotFoundError):
        return "CPU"

def get_system_info():
    """Collects system information needed for the CSV output."""
    # Get CPU info
    cpu_model = platform.processor()
    cpu_count = psutil.cpu_count()

    # Get RAM info
    ram_total_size = psutil.virtual_memory().total / (1024**3)  # in GB

    # Get OS info
    os_info = platform.system() + "-" + platform.release()

    # Get Python version
    python_version = platform.python_version()

    return {
        "cpu_model": cpu_model,
        "cpu_count": cpu_count,
        "ram_total_size": ram_total_size,
        "os": os_info,
        "python_version": python_version
    }

def calculate_ml_carbon_footprint(start_time, end_time, gpu_name=None, carbon_intensity_gco2_kwh=400):
    """Calculates the carbon footprint of a completed ML training run."""
    if gpu_name is None:
        gpu_name = get_gpu_name()
    gpu_power_data = {'A100': 400, 'V100': 250, 'T4': 70, 'P100': 250, 'K80': 300, 'RTX3080': 320, 'RTX3090': 350, 'A10G': 150, 'T1000': 50}
    if gpu_name == "CPU":
        return {"error": "Cannot estimate CPU power consumption."}
    power_consumption_watts = gpu_power_data.get(gpu_name)
    if power_consumption_watts is None:
        return {"error": f"Unknown GPU: {gpu_name}"}
    training_time_seconds = end_time - start_time
    training_time_hours = training_time_seconds / 3600.0
    energy_kwh = (power_consumption_watts * training_time_hours) / 1000
    carbon_emissions_g = energy_kwh * carbon_intensity_gco2_kwh
    results = {"gpu_name": gpu_name, "training_time_hours": round(training_time_hours, 4), "power_consumption_watts": power_consumption_watts, "energy_kwh": round(energy_kwh, 6), "carbon_intensity_gco2_kwh": carbon_intensity_gco2_kwh, "estimated_emissions_gco2": round(carbon_emissions_g, 2)}
    return results

def estimate_footprint_from_time(training_time_hours, gpu_name=None, carbon_intensity_gco2_kwh=400):
    """Estimates the carbon footprint based on a given training time."""
    if gpu_name is None:
        gpu_name = get_gpu_name()
    gpu_power_data = {'A100': 400, 'V100': 250, 'T4': 70, 'P100': 250, 'K80': 300, 'RTX3080': 320, 'RTX3090': 350, 'A10G': 150, 'T1000': 50}
    if gpu_name == "CPU":
        return {"error": "Cannot estimate CPU power consumption."}
    power_consumption_watts = gpu_power_data.get(gpu_name)
    if power_consumption_watts is None:
        return {"error": f"Unknown GPU: {gpu_name}"}
    energy_kwh = (power_consumption_watts * training_time_hours) / 1000
    carbon_emissions_g = energy_kwh * carbon_intensity_gco2_kwh
    results = {"gpu_name": gpu_name, "training_time_hours": training_time_hours, "power_consumption_watts": power_consumption_watts, "energy_kwh": round(energy_kwh, 6), "carbon_intensity_gco2_kwh": carbon_intensity_gco2_kwh, "estimated_emissions_gco2": round(carbon_emissions_g, 2)}
    return results

def generate_csv_data(footprint_results, project_name="GreenLense_Test"):
    """Generates data in the same format as the sample CSV."""
    # Get system information
    system_info = get_system_info()

    # Calculate additional values
    duration_seconds = footprint_results["training_time_hours"] * 3600
    emissions_kg = footprint_results["estimated_emissions_gco2"] / 1000  # Convert g to kg
    emissions_rate = emissions_kg / duration_seconds if duration_seconds > 0 else 0

    # Estimate CPU and RAM power (simplified approach)
    cpu_power = 100  # Estimated average CPU power in watts
    ram_power = 20   # Estimated average RAM power in watts

    # Calculate energy for CPU and RAM (simplified)
    cpu_energy = (cpu_power * footprint_results["training_time_hours"]) / 1000
    ram_energy = (ram_power * footprint_results["training_time_hours"]) / 1000

    # GPU energy
    gpu_energy = footprint_results["energy_kwh"]

    # Total energy
    energy_consumed = cpu_energy + gpu_energy + ram_energy

    # Generate IDs
    run_id = str(uuid.uuid4())
    experiment_id = str(uuid.uuid4())

    # Create the CSV data dictionary
    csv_data = {
        "timestamp": datetime.now().isoformat(),
        "project_name": project_name,
        "run_id": run_id,
        "experiment_id": experiment_id,
        "duration": duration_seconds,
        "emissions": emissions_kg,
        "emissions_rate": emissions_rate,
        "cpu_power": cpu_power,
        "gpu_power": footprint_results["power_consumption_watts"],
        "ram_power": ram_power,
        "cpu_energy": cpu_energy,
        "gpu_energy": gpu_energy,
        "ram_energy": ram_energy,
        "energy_consumed": energy_consumed,
        "water_consumed": 0,  # Default value
        "country_name": "Unknown",  # Default value
        "country_iso_code": "UN",  # Default value
        "region": "Unknown",  # Default value
        "cloud_provider": "",  # Default value
        "cloud_region": "",  # Default value
        "os": system_info["os"],
        "python_version": system_info["python_version"],
        "codecarbon_version": "3.0.8",  # Default value
        "cpu_count": system_info["cpu_count"],
        "cpu_model": system_info["cpu_model"],
        "gpu_count": 1 if footprint_results["gpu_name"] != "CPU" else 0,
        "gpu_model": f"1 x {footprint_results['gpu_name']}" if footprint_results["gpu_name"] != "CPU" else "",
        "longitude": 0,  # Default value
        "latitude": 0,  # Default value
        "ram_total_size": system_info["ram_total_size"],
        "tracking_mode": "machine",
        "on_cloud": "N",
        "pue": 1,  # Default value
        "wue": 0  # Default value
    }

    return csv_data

def write_to_csv(csv_data, filename="emissions.csv"):
    """Writes the data to a CSV file in the same format as the sample."""
    # Define the fieldnames in the same order as the sample CSV
    fieldnames = [
        "timestamp", "project_name", "run_id", "experiment_id", "duration",
        "emissions", "emissions_rate", "cpu_power", "gpu_power", "ram_power",
        "cpu_energy", "gpu_energy", "ram_energy", "energy_consumed", "water_consumed",
        "country_name", "country_iso_code", "region", "cloud_provider", "cloud_region",
        "os", "python_version", "codecarbon_version", "cpu_count", "cpu_model",
        "gpu_count", "gpu_model", "longitude", "latitude", "ram_total_size",
        "tracking_mode", "on_cloud", "pue", "wue"
    ]

    # Check if file exists to determine if we need to write headers
    file_exists = False
    try:
        with open(filename, 'r') as csvfile:
            file_exists = True
    except FileNotFoundError:
        file_exists = False

    # Write to CSV
    with open(filename, 'a', newline='') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        # Write headers if file doesn't exist
        if not file_exists:
            writer.writeheader()

        # Write the data
        writer.writerow(csv_data)

    print(f"Data written to {filename}")

def print_footprint_report(results):
    """Prints a nicely formatted report from the results dictionary."""
    if "error" in results:
        print(f"Calculation failed: {results['error']}")
        return
    print("\n" + "="*50)
    print("         CARBON FOOTPRINT REPORT")
    print("="*50)
    print(f"  GPU Used:                {results['gpu_name']}")
    print(f"  Training Duration:       {results['training_time_hours']} hours")
    print(f"  Avg. Power Consumption:  {results['power_consumption_watts']} Watts")
    print(f"  Total Energy Consumed:   {results['energy_kwh']} kWh")
    print(f"  Grid Carbon Intensity:   {results['carbon_intensity_gco2_kwh']} gCOâ‚‚/kWh")
    print("-"*50)
    print(f"  ğŸŒ ESTIMATED EMISSIONS:  {results['estimated_emissions_gco2']} gCOâ‚‚e")
    print("="*50 + "\n")

print("Carbon calculator functions are ready!")

# --- ESTIMATE FOR YOUR COMPLETED RUN ---

# 1. Make your best guess for the total training time in hours
# Was it 1 hour? 2.5 hours? 3 hours? Let's use 1.0 for this example.
estimated_training_hours = 1.0

# 2. We know you used a T4, so we can specify it for accuracy
# (If you leave it as None, it will auto-detect, which is also fine)
my_gpu = "T4"

# 3. Call the estimation function
print(f"Estimating footprint for a {estimated_training_hours}-hour training run on a {my_gpu}...")

estimated_results = estimate_footprint_from_time(
    training_time_hours=estimated_training_hours,
    gpu_name=my_gpu,
    carbon_intensity_gco2_kwh=400 # You can adjust this value
)

# 4. Print the report
print_footprint_report(estimated_results)

# 5. Generate CSV data and write to file
csv_data = generate_csv_data(estimated_results)
write_to_csv(csv_data)

# 6. Print a preview of the CSV data
print("\nCSV Data Preview:")
for key, value in csv_data.items():
    print(f"{key}: {value}")
```
